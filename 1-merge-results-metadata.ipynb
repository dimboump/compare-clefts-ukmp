```python
import os
import re
import pandas as pd
```

```python
COLS = ['nr', 'pre', 'match-2', 'match-1', 'match', 'match+1', 'match+2', 'post', 'id']
DIR = r"C:\Users\dimboump\workspace\compare-clefts-ukmp\data"
os.chdir(DIR)
```

```python
with open("antconc_results_hansard.txt", 'r', encoding="utf8") as sfile:
	with open("temp.csv",'w+', encoding="utf8") as tfile:
		# Provide line with headers for the modified results file
		# Make sure the column containing the file ids has the same column name as it has in the metadata - in this case 'id'
		tfile.writelines("\t".join(COLS) + "\n")
		lines = sfile.readlines()

		for line in lines:
			# replace whitespace by tabs for the 3 words preceding and following the first word of the match
			pattern = re.compile("(\w+\W+)(\w+\W+)(\w+\W+)\t(\w+\W+)(\w+\W+)(\w+\W+)(\w+\W+)")
			line = re.sub(pattern, "\g<1>\t\g<2>\t\g<3>\t\g<4>\t\g<5>\t\g<6>\t\g<7>", line)

			# replace filename extension by nothing, so that the last column is identical to the ID column of the metadata file
			pattern = re.compile(r".txt")
			line = re.sub(pattern, "", line)

			tfile.writelines(line)
```

```python
temp_df = pd.read_csv("temp.csv", encoding="utf8", sep='\t', engine='python')
meta_df = pd.read_csv('HansardMetaDataTexts_fixed.csv', encoding='utf8', sep='\t', engine='python')
# meta_df['id'] = meta_df['id'].astype(str)
```

```python
merged_df = pd.merge(temp_df, meta_df, on='id')
merged_df = merged_df.rename(columns={'display_as': 'name'})
merged_df.sample(5)
```

```python
authors_df = pd.read_csv('HansardMetaDataAuthors.csv', encoding='utf8', sep='\t', usecols=['name', 'year_of_birth','gender'], engine='python')
```

```python
print(merged_df.shape)
merged_df = pd.merge(merged_df, authors_df, on='name')
print(merged_df.shape)
```

```python
merged_df = (merged_df
    .assign(person_id=merged_df['url'].str.split('/').str[-1].astype(int))
    .drop(['oral_heading'], axis=1)
    .astype({
        'party': 'category',
        'constituency': 'category',
        'hansard_membership_id': 'int16',
        'speakername': 'category',
        'speech_class': 'category',
        'gender': 'category'})
)
```
