## Cross-corpus analysis

### Cleft properties

#### Type

```r
han_type <- table(han$type, exclude=c("perf", "other"))
bnc_type <- table(bnc$type, exclude=c("ambiguous", "other"))

type_matrix = matrix(c(han_type, bnc_type), nrow=2, ncol=3)

rownames(type_matrix) <- c("MPs", "BNC")
colnames(type_matrix) <- c("canonical", "IP", "unclear")

mosaicplot(type_matrix, main="Cleft type per corpus", color=c("#a52040", "#404080", "#6060c0"), border=FALSE, ylab="Type")
```

#### Givenness

```r
han_given <- table(han$givenness, exclude=c("UNK"))
bnc_given <- table(bnc$givenness, exclude=c("un"))

given_matrix = matrix(c(han_given, bnc_given), nrow=2, ncol=4, byrow=TRUE)

rownames(given_matrix) <- c("MPs", "BNC")
colnames(given_matrix) <- c("active", "inactive", "new", "part")

mosaicplot(given_matrix, main="Cleft givenness per corpus", color=c("#a52040", "#404080", "#6060c0", "#8080fc"), border=FALSE, ylab="Givenness")
```

#### Factuality

```r
han_fact <- table(han$factuality, exclude=c("other"))
bnc_fact <- table(bnc$factuality)

fact_matrix = matrix(c(han_fact, bnc_fact), nrow=2, ncol=3)

rownames(fact_matrix) <- c("MPs", "BNC")
colnames(fact_matrix) <- c("fact", "non-fact", "unclear")

mosaicplot(fact_matrix, main="Cleft factuality per corpus", color=c("#a52040", "#404080", "#6060c0"), border=FALSE, ylab="Factuality")
```

#### Focus

```r
han_focus <- table(han$focus, exclude=c("stranded"))
bnc_focus <- table(bnc$focus)

focus_matrix = matrix(c(han_focus, bnc_focus), nrow=2, ncol=3, byrow=TRUE)

rownames(focus_matrix) <- c("MPs", "BNC")
colnames(focus_matrix) <- c("adv", "clause", "NP")

mosaicplot(focus_matrix, main="Cleft focus per corpus", color=c("#a52040", "#404080", "#6060c0"), border=FALSE, ylab="Focus")
```

### Associations

```r
library(vcd)

assoc(type_matrix, shade=TRUE)
```

```r
assoc(given_matrix, shade=TRUE)
```

```r
assoc(fact_matrix, shade=TRUE)
```

```r
assoc(focus_matrix, shade=TRUE)
```
